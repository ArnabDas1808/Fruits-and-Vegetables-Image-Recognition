```python
import os
from PIL import Image

# Update dataset path
dataset_path = '/content/Fruits-by-YOLO-1/train/images'

sizes = []
if os.path.exists(dataset_path):
    for filename in os.listdir(dataset_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            with Image.open(os.path.join(dataset_path, filename)) as img:
                sizes.append(img.size)  # Collect (width, height)
    print("Image sizes in dataset:", sizes)
else:
    print(f"Error: Path '{dataset_path}' does not exist.")
```
